```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
```

```python
# Import the combined dataset
df_combined = pd.read_csv("combined_data.csv")
```

```python
df_combined
```

```python
df_combined.info()
```

```python
# Import airports dataset
df_airports = pd.read_csv("airports.csv")
```

```python
df_airports.head(10)
```

```python
df_airports.info()
```

```python
# Drop unnecessary columns
df_drop = df_combined.iloc[:, [0,1,16,17]]
df_drop
```

```python
# Merge the combined dataset with the airports dataset
df_merged = pd.merge(df_drop, df_airports, left_on= 'Origin', right_on= 'iata', how = 'inner')
```

```python
df_merged
```

```python
df_merged = pd.merge(df_merged, df_airports, left_on= 'Dest', right_on= 'iata', how = 'inner')
```

```python
df_merged.head()
```

```python
df_merged.info()
```

```python
# Remove unnecessary columns
df_merged = df_merged.iloc[:, [0,1,7,14]]
df_merged.head()
```

```python
# Rename columns
df_merged = df_merged.rename(columns = {"state_x" : "Origin_state", "state_y" : "Dest_state"})
df_merged.head()
```

```python
# Remove rows with no values
df_final = df_merged.dropna(subset = ["Origin_state","Dest_state"])
```

```python
df_final
```

```python
# Group by Year and then month and get the count
df_group = df_final.groupby(['Year','Month']).count()
```

```python
df_group
```

## Line plot to compare number of flights

```python
# Create a line plot with month in the x axis and number of flights in y axis

g1 = sns.lineplot(data = df_group, x = "Month", y = "Origin_state", hue = "Year", palette = ['black','red'])
g1.set_xticks(range(1,13))
g1.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
g1.set_ylabel("Number of Flights")
g1.set_title("Line plot showing number of flights per month in 2006 and 2007")
```

Since there is no information of passengers, number of flights is taken. Assume that all flights are full

Number of flights has increased in 2007 when compared to 2006

February has the least flights, then there is a steady rise which peaks at August

## Find most used flight routes

```python
df_final.dtypes
```

```python
# Create new column to combine origin and dest columns
df_final['Origin:Dest'] = df_final['Origin_state'].astype(str) + ":" + df_final['Dest_state'].astype(str)
```

```python
df_final
```

```python
# Keep only year and origin:dest columns
df_origin_dest = df_final.iloc[:, [0,-1]]
```

```python
df_origin_dest
```

```python
# Create a new column to count each flight 
df_origin_dest['Number of People'] = 1
```

```python
df_origin_dest
```

```python
# Groupby Origin:Dest and get the sum
df_no_of_people = df_origin_dest.groupby(['Year','Origin:Dest']).sum().reset_index()
df_no_of_people
```

## Split the dataset into 2 by year

```python
df_count_2006 = df_no_of_people[df_no_of_people['Year'] == 2006]
df_count_2006
```

```python
df_count_2007 = df_no_of_people[df_no_of_people['Year'] == 2007]
df_count_2007
```

## Sort the new datasets in descending order

```python
df_count_2006 = df_count_2006.sort_values(by=['Number of People'], ascending=False)
df_count_2006.head()
```

```python
df_count_2007 = df_count_2007.sort_values(by=['Number of People'], ascending=False)
df_count_2007.head()
```

The tables shows the most and least frequently used routes

In 2006, the top 5 routes are CA:CA, TX:TX, HI:HI, CA:AZ and AZ:CA

In 2007, the top 5 routes are CA:CA, TX:TX, HI:HI, AZ:CA and CA:AZ
