```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
```

```python
# Import the 2 datasets. We are using the datasets from 2006 and 2007
df_2006 = pd.read_csv("2006.csv")
```

```python
df_2007 = pd.read_csv("2007.csv")
```

```python
# Combine the datasets
df_combined = pd.concat([df_2006, df_2007], ignore_index=True)

# Export the combined dataset which will be used in the following questions
df_combined.to_csv('combined_data.csv', index = False)
```

```python
# Check number of null values in each column
df_combined.isna().sum()
```

```python
# Remove rows that don't contain departure or arrival times
df_drop = df_combined.dropna(subset=['DepTime', 'ArrTime'])
```

```python
df_drop.isna().sum()
```

```python
df_drop.shape
```

```python
# Convert ArrDelay and DepDelay columns from float to int
df_drop = df_drop.astype({"ArrDelay":"int","DepDelay":"int"})
```

```python
# Add Arrival Delay and Departure delay to get total delay and create a new column to store total delays
df_drop['TotalDelay'] = df_drop['ArrDelay'] + df_drop['DepDelay']
```

```python
# Create timebins with each bin having a width of 3 hours using the CRSDepTime column
df_drop['Timebins'] = pd.cut(df_drop['CRSDepTime'], [0,300,600,900,1200,1500,1800,2100,2400])
```

```python
df_drop.head()
```

```python
# Remove columns which are unnecessary for this question
df_final = df_drop.iloc[:, [1,2,3,5,14,15,29,30]]
```

```python
df_final # This is the cleaned dataset which will be used to answer each part of this question
```

## When is the best time of the day to fly to minimise delays?

```python
# Remove unnecessary columns
df_time_of_day = df_final.iloc[:, 4:]
```

```python
df_time_of_day.head()
```

```python
# Group by Timebins and calculate the median
df_time_of_day_grouped = df_time_of_day.groupby('Timebins').median().reset_index()
df_time_of_day_grouped
```

```python
# Side by side boxplots for each timebin. Triangle shows the mean

sns.set(rc={"figure.figsize":(8, 6)})
g1 = sns.boxplot(data = df_time_of_day, x = "TotalDelay", y = "Timebins", color = "yellow",
                 showmeans = True,
                 flierprops ={ 'marker': 'o','markeredgecolor': 'blue', 'markerfacecolor': 'white'},
                 medianprops = {'color': 'black'},)
g1.set_title("Horizontal boxplots of Total Delay by Time of day")
g1.set_xlim(-100,100)
```

03:00 - 06:00 is the best time to fly to minimise delays

## When is the best day of the week to fly to minimise delays?

```python
# Remove unnecessary columns
df_day_of_week = df_final.iloc[:, [2,-2]]
df_day_of_week
```

```python
# Group by day of week and get the mean delays
avg_delay = df_day_of_week.groupby(["DayOfWeek"]).mean().reset_index()
avg_delay.rename(columns={"TotalDelay": "AverageDelay"})
```

```python
# Create a line plot with day on x axis and delay on y axis

g2 = sns.barplot(data = avg_delay, x = "DayOfWeek", y = "TotalDelay")
g2.set_xlabel("Day of Week")
g2.set_ylabel("Average Delay(min)")
g2.set_title('Bar plot showing average delay for each day')

g2.set_xticks(range(0,7))
g2.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
```

Saturday is the best day of the week to fly to minimise delays

## When is the best time of the year to fly to minimise delays?

```python
df_final
```

```python
# Remove unnecessary columns
df_time_of_year = df_final.iloc[:, [0,-2]]
```

```python
df_time_of_year
```

```python
# Group by each month to find the average delay
month_avg_delay = df_time_of_year.groupby(["Month"]).mean()
month_avg_delay.rename(columns={"TotalDelay": "AverageDelay"})
```

```python
# Create a line plot with month on x axis and delay on y axis

sns.set_style ("ticks")
g3 = sns.lineplot(data = month_avg_delay, x = "Month", y = "TotalDelay" , color = 'red')
g3.set_ylabel("Average Delay(min)")
g3.set_title('Line plot showing average delay for each month')
g3.set_xticks(range(1,13))
g3.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
```

September and November are the best months of the year to fly to minimise delays
